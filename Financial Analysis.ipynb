### Time Series Components

Seasonality-Zaman Serileri 3 katmandan oluşur. Mevsimsellik, Trend ve Irregular katmanları orjinal seriyi verir.

Mevsimsellik : İklim, insan alışkanlıkları, sosyal/siyasal/ekonomik dalgalanmalar gibi değişkenliklerin belirli zaman dilimlerinde seri üzerinde yaptığı örüntülerdir diyebiliriz.
Örnek olarak köpek balığı saldırlarını içeren bir zaman serisi olduğunu düşünelim 5 yıllık olsun. Bu zaman serisi yazları artacaktır ve kışları azalacaktır çünkü insanlar yaz aylarında denize girdikleri için.

Trend : Serilerdeki uzun süreli artış veya azalışları yansıtmaktadır. Çok önemli bir componentdir.
Örnek olarak şuan pandemi sürecini yaşıyoruz ve aşağı, yukarı eğriler olacağını söyleyebiliriz.

Öngörülemeyen Kısım : Serilerdeki trend ve mevsimsellikten arındırıldıktan sonra kalan kısımdır diyebiliriz. Bu dalgalanmalar öngörülemez, kontrol edilemez ve düzensiz hareketlerdir. 
Pandemi Dönemini geçiriyoruz, doğanın verdiği bir etki bu etkiyi modellemek veya tahminlemek mümkün olmadığı için. Bu etki yani istatiksel olarak modellenemez.

### Hareketli Ortalama Yöntemi
#### Çok temel bir yöntem olduğunu söyleyebilirim, teknik analiz ile al sat tavsiyeleri oluşturabiliriz.
#### Basit hareketli ortalama = kapanış fiyatının toplam değerlerinin zaman aralığına bölünmesiyle hesaplanır. Mesela 5 günlük ortalama hesaplanırken, 5 günün kapanışları toplanarak 5 e bölünür.

```python
(len(thy_df)//2) # bu şekilde datayı yarıya indirebiliriz, daha iyi görselleştirme için
```

```python
close_thy = thy_df.iloc[(len(apple_df)//2):,3]
close_thy.head()
```

```python
rolling_app5 = close_thy.rolling(window=5).mean() # # rolling = yuvarlama
rolling_app14 = close_thy.rolling(window=14).mean()
rolling_app21 = close_thy.rolling(window=21).mean()

# aldığımız close değerlerini 5windows, 14window, 21windows olarak aldık ve ortalamasını aldık
```

```python
MAs = pd.concat([close_thy, rolling_app5, rolling_app14, rolling_app21], axis=1)
MAs.columns = ['Close', 'Close_5', 'Close_14', 'Close_21']
MAs.head(30)

# Close feature'u herhangi bir işlem uygulanmamış feature
# Close_5 ise 5 günlük hareketli ortalaması demek oluyor aslında
# böyle kaydırarak 5 günlük, 14 günlük ve 21 günlük veri seti elde etmiş oluyoruz, aslında bilgi verici bir durum diyebiliriz
```

```python
fig, ax = plt.subplots(figsize=(15,10))
ax.plot(close_thy.index, close_thy, label='THY')
ax.plot(rolling_app5.index, rolling_app5, label='5 days rolling')
ax.plot(rolling_app14.index, rolling_app14, label='14 days rolling')
ax.plot(rolling_app21.index, rolling_app21, label='21 days rolling')
ax.legend(loc='upper left')
# bu verisetini görselleştirerek aslında kendimize al sat tavsiyeleri verebiliriz, bu ortalamalar değerlerinin kesişim noktalarına bakarak
# yukarıda datayı ikiye bölerek yaparsak daha iyi okuma yapabiliriz 

# ikinci kesişmeye bakarsak, üçüncü kesişmeye kadar kırmızı çizgi alt tarafta olduğu için artış trendi devam ediyor 
# aslında graph'ın özeti şu kırmızı çizginin aşağıda olması bize artış trendinin devam ettiğini söylüyor
# kırmızı, yeşil, turuncu çizgilerin birbirleri ile hareketliğine bakarak biz al,sat tavsiye verebiliriz
# bu otomatizasyonu uyguladığımız zaman bile sadece hareketli ortalama yöntemi ile birçok indükatörün yapamadığını yapabiliriz.
# İndikatörler forex piyasasında teknik olarak analizler yaparken yatırımcılara yardımcı olabilen teknik bir göstergelerdir.
# İndikatörler incelenerek piyasa hakkında yön tahminleri yapabilir ve çizilmiş olan grafik verileriyle desteklenebilir.
```
